```python
def extractFrames(pathIn, pathOut, start_time=5000, end_time=120000, time_step=2000):
    os.mkdir(pathOut)
 
    cap = cv2.VideoCapture(pathIn)
    count = 0
 
    for k in range(start_time, end_time+1, time_step):
        # Capture frame-by-frame
        cap.set(cv2.CAP_PROP_POS_MSEC, k)
        success, frame = cap.read()
 
        if success == True:
            print('Read %d frame: ' % count, success)
            frame = cv2.resize(frame,(frameWidth, frameHeight))
            cv2.imwrite(os.path.join(pathOut, "frame{:d}.jpg".format(count)), frame)  # save frame as JPEG file
            count += 1
        else:
            print(success, frame)
            break
 
    # When everything done, release the capture
    cap.release()
    cv2.destroyAllWindows()
 
extractFrames('Dataset/B.mp4', 'data')
```

```python
def saveFrames(directory):
    for file_num, file_name in enumerate(os.listdir(directory)):
        file_path = directory + '/' + file_name
        save_frames_path = directory + '/' + str(file_num)
        extractFrames(file_path, save_frames_path)
```

```python
saveFrames('Dataset')
```
